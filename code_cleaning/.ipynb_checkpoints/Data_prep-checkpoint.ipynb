```python
#import modules for analyzing ,plotting, and formatting
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import seaborn as sns
import sqlite3
import os
import folium
```

```python
os.chdir(r"C:\Users\5luca\Documents\Python\Projects\wildfire_ca\data")
df = pd.read_csv('California_Fire_Incidents.csv')
weather = pd.read_csv('weather_ca.csv')
df.head()
```

```python
selected_columns = [
    'AcresBurned',
    'ArchiveYear',
    'Counties',
    'Latitude',
    'Longitude',
    'MajorIncident',
    'Name',
]
```

```python
df2 = df[selected_columns].copy()
df2
```

```python
weather['StartTime(UTC)']= pd.to_datetime(weather['StartTime(UTC)'])
weather['EndTime(UTC)']= pd.to_datetime(weather['EndTime(UTC)']) 

weather['StartTime(UTC)'].info()

weather_2016 = weather[(weather['StartTime(UTC)'].dt.year == 2016) & (weather['EndTime(UTC)'].dt.year == 2016)]
weather_2016 =  weather_2016.rename(columns={"Precipitation(in)": "Precipitation"})
weather_2016 
```

```python
df_select = df2[df2.Longitude<-115]
df_select = df_select[(df_select.Latitude<44) & (df_select.Latitude > 30)]
plt.scatter(df_select.Longitude, df_select.Latitude)
plt.grid()
plt.show()
```

```python
#Majordf = df_select.iloc[i]['Longitude'== 'True']]
#map
zoom_factor = 5 
radius_scaling = 25

ca_map = folium.Map (location = [36,-120], zoom_start = zoom_factor)

for i in range (0,df_select.shape[0]):
    folium.Circle(
      location=[df_select.iloc[i]['Latitude'], df_select.iloc[i]['Longitude']],
      radius=np.sqrt(df_select.iloc[i]['AcresBurned'])*radius_scaling,
      color='red',
      fill=True,
      fill_color='red'
   ).add_to(ca_map)
ca_map
```

```python
import plotly.express as px

color_scale = [(0, 'orange'), (1,'red')]
color_scale1 = [(0, 'teal'), (1,'blue')]

fig = px.scatter_mapbox(df_select, 
                        lat= "Latitude", 
                        lon= "Longitude", 
                        color= "MajorIncident",
                        category_orders = {'FIRE_SIZE_CLASS': ['False','True']},
                        color_discrete_sequence=px.colors.qualitative.G10,
                        #size= "AcresBurned",
                        opacity=0.7,
                        zoom=5.5, 
                        height=600,
                        width=600)

fig.add_trace(px.scatter_mapbox(weather_2016,
                                lat="LocationLat",
                                lon="LocationLng",
                                color ='Type',
                                category_orders = {'Type': ['Fog', 'Rain', 'Cold',
                                                            'Snow', 'Storm', 'Precipitation', 
                                                            'Hail']},
                                color_continuous_scale=px.colors.qualitative.T10,
                                #size ='Precipitation(in)',
                                opacity=0.1
                                ).data[0])

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(legend=dict(
    orientation="h",
    yanchor="bottom",
    y=1.02,
    xanchor="right",
    x=1
))
fig.update_layout(legend_tracegroupgap=2)
fig.update_layout(showlegend=True)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
